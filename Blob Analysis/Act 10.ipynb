```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np 
import cv2

plt.style.use('dark_background')
```

```python
#ThresholdingandMorphing for 2 lines:
def morph1(image_sub1,struct1):
    G_line1 = cv2.cvtColor(image_sub1, cv2.COLOR_RGB2GRAY)
    thresh1, BW1 = cv2.threshold(G_line1, 200, 255, cv2.THRESH_BINARY_INV)
    cl_morph1 = cv2.morphologyEx(BW1, cv2.MORPH_OPEN, struct1)
    op_morph1 = cv2.morphologyEx(cl_morph1, cv2.MORPH_OPEN, struct1)
    return op_morph1

def morph(image_sub1,struct1,thresh):
    G_line1 = cv2.cvtColor(image_sub1, cv2.COLOR_RGB2GRAY)
    thresh1, BW1 = cv2.threshold(G_line1, thresh, 255, cv2.THRESH_BINARY_INV)
#    op_morph1 = cv2.morphologyEx(BW1, cv2.MORPH_OPEN, struct1)
#    return [BW1,op_morph1]
    return BW1
```

```python
Blob = cv2.cvtColor(cv2.imread('crop_blobs.jpg'), cv2.COLOR_BGR2RGB)  
struct = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2,1))
BB = morph(Blob,struct,200)
thresh1, BW1 = cv2.threshold(Blob, 200, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize=[15,7])
plt.rcParams['axes.titlesize'] = 174
plt.subplot(121)
plt.imshow(Blob)
plt.subplot(122)
plt.imshow(BW1,'gray')
plt.yticks(np.arange(0,250,25))
plt.show()
plt.figure(figsize=[15,7])
plt.rcParams['axes.titlesize'] = 174
plt.subplot(121)
plt.imshow(BB,'gray')
plt.subplot(122)
plt.imshow(BB,'gray')
plt.show()
```

```python
# Read image
im = cv2.imread("crop_blobs.jpg", cv2.IMREAD_GRAYSCALE)
 
# Set up the detector with default parameters.
detector = cv2.SimpleBlobDetector_create()
 
# Detect blobs.
keypoints = detector.detect(im)
 
# Draw detected blobs as red circles.
# cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS ensures the size of the circle corresponds to the size of blob
im_with_keypoints = cv2.drawKeypoints(im, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
 
# Show keypoints
cv2.imshow("Keypoints", im_with_keypoints)
#cv2.waitKey(0)
```
